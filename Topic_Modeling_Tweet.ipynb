## Read-in list of tweets

First one needs to read-in tweets. It is not too difficult with Twitter API and the python packages. I largely did it in similar fashions as this tutorial here. https://realpython.com/twitter-bot-python-tweepy/. One can easily get one's own key from twitter API and download any public tweets one wants. 

Here I have already downloaded the tweets and saved them as tsvs and I am just loading the downloaded tweets. 

```python
Tweet_list = ["twitter_handle1", "twitter_handel2"]
```

```python
def read_in_tweets(handle): 
    contents_all = []
    with open("../" + handle + ".tsv") as f: 
        for l in f.readlines():
            l = l.split('\t')
            if l[0] != 'id':
                contents_all.append(l[-1])
    return(contents_all)
```

## Text cleaning

Stop words are shallow words that are typically removed in semantic analyses, although they can sometimes contain information about style. Here I will just simply remove those because I am more focused on topic extraction

```python
stop_words = set(stopwords.words('english'))
```

Next I did some basic text cleaning removing @s, tags, puncutaion, etc.

```python
### Remove punctuation for unicode strings (code I got from online)
import regex as re

def remove_punctuation(text):
    return re.sub(u"\p{P}+", "", text)


def data_clean(line):
    l = []
    line = line.split()
    for i in line:
        i = i.lower()
        if i[0] != '@' and i[0] != "#": # Getting rid of hashtag and @s
            i = remove_punctuation(i)
            if len(i) != 0 and i[:5] != 'https' and i not in stop_words and i != 'rt' and i != 'amp': # Getting rid of 
                l.append(i)
    return l
```

All the tweets of each twitter handle is now stored in a dictionary.

```python
cleaned_tweets = dict()

for handle in Tweet_list:        
    cleaned_tweets[handle] = [data_clean(l) for l in read_in_tweets(handle)]
```

## Topic Modeling

Next, I use gensim package to extract topics from the tweets. 

```python
from gensim import corpora
import pickle
import gensim
import copy
```

I do topic modeling separately for each twitter account and treat each tweet as a document. 

```python
def get_topics(handle, NUM_TOPICS, NUM_WORDS, NUM_PASSES = 15, seed = 123):
    dictionary = corpora.Dictionary(cleaned_tweets[handle])
    corpus = [dictionary.doc2bow(text) for text in cleaned_tweets[handle]]
    # pickle.dump(corpus, open('corpus/' + handle + '_corpus.pkl', 'wb'))
    # dictionary.save('dictionary/' + hanlde + '_dictionary.gensim')
    
    ldamodel = gensim.models.ldamodel.LdaModel(corpus, num_topics = NUM_TOPICS, id2word=dictionary, passes = NUM_PASSES, random_state = seed)
    ## Getting the topics with the top words asscociated with this topic
    topics = ldamodel.print_topics(num_words=NUM_WORDS)
    
    with open(handle + '_model_' + str(NUM_TOPICS) + '.pkl', 'wb') as f:
        pickle.dump(ldamodel, f, pickle.HIGHEST_PROTOCOL)    
    
    topics_dic = {}
    
    for i in range(NUM_TOPICS):
        topic = topics[i][1].split("+")
        topic_dic = {}
        for item in topic:
            weight = float(item.split("*")[0])
            key = item.split("*")[1].split("\"")[1]
            topic_dic[key] = weight
        topics_dic[i] = topic_dic.copy()
        
    return(topics_dic)

def print_topics(topic_dic):
    for i in topic_dic.keys():
        print("Topic " + str(i + 1) + ":")
        for j in topic_dic[i].keys():
            print('  ' + j + ' (' + str(topic_dic[i][j]) + ')')    
```

There are many parameters that one can fine tune. Here I am telling the model to extract 5 topics from all the tweets. One can also tweak the number to see which ones work better. 

I also specify how many words I want to display when I print out the topics. 

```python
## Hyper-parameter for training and for display
num_topic = 5
num_words = 10
```

```python
handle = "handle1"
H1_dic = get_topics(handle, num_topic, num_words)
with open(handle + '_dic_' + str(num_topic) + '.pkl', 'wb') as f:
    pickle.dump(H1_dic, f, pickle.HIGHEST_PROTOCOL)
```

## Creating 'word' embeddings for each topic

Here I am using glove features to generate word embeddings (https://nlp.stanford.edu/projects/glove/). I used the one trained on twitter data for obvious reasons. 

```python
with open("../glove_dict.pkl" , "rb") as f:
    glove_dic = pickle.load(f)
```

For each topic, I just calcuated a weighted average of the words that are associated with the topic. Here I am only using the top K words spcified above with the NUM_Words parameter

```python
vocab = glove_dic.keys()

def create_embeddings(topic_dic):
    vec_dic = {}
    for key in topic_dic.keys():
        topic = topic_dic[key]
        acc = 0
        acc_vec = np.zeros(100)
        for key_word in topic: 
            if key_word in vocab:
                acc_vec += glove_dic[key_word] * topic[key_word]
                acc += topic[key_word]
        if acc == 0:
            print("No word found in the vocabulary :(")
        else:
            acc_vec /= acc
        vec_dic[key] = acc_vec
    return(vec_dic)
```

```python
H1_vec = create_embeddings(H1_dic)
handle = "twitter_handle1"
with open(handle + '_vec_' + str(num_topic) + '.pkl', 'wb') as f:
    pickle.dump(H1_vec, f, pickle.HIGHEST_PROTOCOL)
```
